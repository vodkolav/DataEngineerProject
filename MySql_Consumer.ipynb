```python
from kafka import KafkaProducer
from kafka import KafkaConsumer
import pyarrow as pa
from datetime import datetime
import json
from time import sleep

import mysql.connector as mc
import datetime

import csv
import re

# MySQL
mysql_port = 3306
mysql_database_name = 'Twitter'
mysql_table_name = 'src_events'
mysql_username = 'naya'
mysql_password = 'naya'
host = 'localhost'
# # Topics/Brokers
topic1 = 'TweeterArchive'
brokers = ['localhost:9092']



```

```python
mysql_conn = mc.connect(
    user = mysql_username,
    password = mysql_password,
    host = host,
    port = mysql_port,
    autocommit=True)
```

```python
mysql_cursor = mysql_conn.cursor()
```

```python
insert_statement = """
    INSERT INTO Twitter.Tweet(id, text) 
    VALUES ('{}', '{}');"""
```

```python
def mysql_tweet_insert(mysql_conn, id, text):
    mysql_cursor = mysql_conn.cursor()
    print (insert_statement.format(id, text))
    mysql_cursor.execute(insert_statement.format(id, text))
    mysql_cursor.close()
```

```python
consumer = KafkaConsumer(
    topic1,
    group_id = 'MySQL',
    bootstrap_servers = ['localhost:9092'],
    auto_offset_reset = 'largest',
    enable_auto_commit = True,
    auto_commit_interval_ms = 1000)
```

```python
for i, message in enumerate(consumer):
    events = json.loads(message.value)
    id =  str(events['user_id']) 
    text =  '"' + events['text'] + '"'
    text = text.replace("\'", "")
    mysql_tweet_insert(mysql_conn, id, text)
```
